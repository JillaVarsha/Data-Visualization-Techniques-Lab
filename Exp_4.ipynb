```python
import pandas as pd

df = pd.read_csv("C:/Users/jvars/Downloads/netflix_titles.csv")

print(df.head())
```

```python
print(df.shape)
```

```python
print(df.dtypes)
```

```python
recent_titles = df[df["release_year"] > 2020]

print(recent_titles)
```

```python
us_titles = df[df["country"] == "United States"]

print(us_titles)
```

```python
sorted_df = df.sort_values(
    by="release_year",
    ascending=False
)

print(sorted_df)
```

```python
print("Mean =",
      df["release_year"].mean())

print("Median =",
      df["release_year"].median())

print("Mode =",
      df["release_year"].mode()[0])

print("Standard Deviation =",
      df["release_year"].std())
```

```python
print(df.describe())
```

```python
type_avg = df.groupby(
    "type"
)["release_year"].mean()

print(type_avg)
```
