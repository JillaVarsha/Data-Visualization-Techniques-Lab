```python
#creating database
import pandas as pd
import sqlite3

df = pd.read_csv("/Users/student/Downloads/netflix_titles.csv")

conn = sqlite3.connect("netflix.db")

df.to_sql("netflix_titles", conn, if_exists="replace", index=False)

conn.close()

print("SQLite database created successfully!")
```

```python
#importing from database
import sqlite3
import pandas as pd

conn = sqlite3.connect("netflix.db")
print('connection successful')

df = pd.read_sql_query("SELECT * FROM netflix_titles;", conn)

print(df.head())

conn.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/student/Downloads/netflix_titles.csv")

type_count = df['type'].value_counts()

plt.figure(figsize=(6,4))
plt.bar(type_count.index, type_count.values)
plt.title("Distribution of Netflix Content")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/Users/student/Downloads/netflix_titles.csv")

genre_count = df['listed_in'].str.split(', ').explode().value_counts().head(10)

plt.figure(figsize=(10,5))
plt.bar(genre_count.index, genre_count.values)
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```
