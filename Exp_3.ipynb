```python
import pandas as pd
df = pd.read_csv("C:/Users/jvars/Downloads/netflix_titles.csv")
print(df.isnull())
```

```python
print(df.isnull().sum())
```

```python
columns = ["director", "cast", "country", "date_added", "rating", "duration"]

for col in columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print(df)
```

```python
print(df.duplicated())

df.drop_duplicates(inplace=True)

print(df)
```

```python
Q1 = df['release_year'].quantile(0.25)
Q3 = df['release_year'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['release_year'] < lower) |
              (df['release_year'] > upper)]

print(outliers)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

num_cols = df.select_dtypes(include=['int64', 'float64']).columns

df[num_cols] = scaler.fit_transform(df[num_cols])

print(df[num_cols])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df['rating'] = encoder.fit_transform(df['rating'])

print(df)
```
